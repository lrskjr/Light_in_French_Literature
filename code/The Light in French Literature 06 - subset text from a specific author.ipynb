```python
import re
import os
import pandas as pd
from pathlib import Path
```

```python
input_dir = Path.cwd() / '../data/csv_files' # input directory
df = pd.read_csv(input_dir /'text_data28_09_24.csv', sep='|')
```

```python

all_zola = ' '.join(df[df['author'] == 'Zola']['text'])
all_balzac = ' '.join(df[df['author'] == 'Balzac']['text'])

output_dir = Path.cwd() / '../data' # input directory
with open(output_dir / 'all_of_zola.txt', 'w', encoding='utf-8-sig') as f:
    f.write(all_zola)
    
with open(output_dir / 'all_of_balzac.txt', 'w', encoding='utf-8-sig') as f:
    f.write(all_balzac)
```

```python
# Group by 'author' and join the text strings using a space as a separator
grouped_df = df.groupby('author')['text'].agg(' '.join).reset_index()

# 'grouped_df' now contains the joined text strings for each author
grouped_df
```
